# Importing necessary libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('SampleSuperstore.csv')
```

# Exploring the dataset

# Checking correlation between columns

```python
sns.heatmap(df1.corr(),annot=True)
plt.show()
```

# As the profit and discount columns have negative correlation which gives us insight that both are inversely proportional which means when discount increases profit decreases and vice versa

# Visualizing data and finding the trends of sales and profit

```python
sns.pairplot(df1)
plt.show()
```

```python
 df1.groupby(['Category','Sub-Category'])['Sales'].sum()#grouping the data and checking total sales for each sub category
```

```python
sns.stripplot(df1['Category'],df1['Sales'])
plt.show()
```

```python
df1.groupby('Sub-Category')['Discount'].sum().plot(kind='bar',color='black')

plt.show()
```

```python
x =  df1.groupby(['Category','Sub-Category'])['Sales'].sum()
plt.figure(figsize=(20,4))
x.plot(kind='bar',color = 'red')
plt.show()
```

Here we get the insight that the maximum sales is given by 'Phones' under Technology category
and minimum sales is given by Fasteners under Office Supplies

```python
df1.groupby(['Sub-Category'])['Profit'].sum()
```

```python
b = df1.groupby(['Sub-Category'])['Profit'].sum()
plt.figure(figsize=(20,6))
sns.barplot(b.index,b.values)
plt.show()
```

Here we get the insight that the maximum profit is given by 'Copiers' under Technology category 
and maximum loss is given by 'Tables' under furniture category.

```python
plt.figure(figsize=(15,10))
df1.groupby('Sub-Category')['Quantity'].sum().plot(kind='pie',autopct = '%.1f%%')
plt.show()
```

Maximum quantities were sold for binders and minimum for copiers

# Checking the sales trend state wise

```python
state_profit = df1.groupby('State')['Profit'].sum().sort_values()
plt.figure(figsize=(15,10))
state_profit.plot(kind='bar',color = 'red')
plt.show()
```

Maximum profit is generated by the state of california and maximum loss occurs in the state of texas

```python
state_profit = df1.groupby('State')['Sales'].sum().sort_values()
plt.figure(figsize=(15,10))
state_profit.plot(kind='bar',color = 'green')
plt.show()
```

Maximum sales is generated by the state of 'California' and minimum sales by 'North dakota'

```python
fig,axes = plt.subplots(1,3,figsize=(20,5))
sns.countplot(df1['Region'],ax=axes[0])
sns.countplot(df1['Ship Mode'],ax=axes[1])
sns.countplot(df1['Segment'],ax=axes[2])
plt.show()
```
